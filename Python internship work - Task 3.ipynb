```python
# Task 3 = Password Generator # 

import random # importing random to generate random characters
import string # it provide ready-made characters

print("=================================")
print("      PASSWORD GENERATOR")
print("=================================")

# User Input

length = int(input("Enter Password Length: "))

include_numbers = input("Include Numbers? (Y/N): ").upper() # .upper = is used to convert the user input to uppercase letters
include_symbols = input("Include Symbols? (Y/N): ").upper()

# Add Basic Characters

characters = string.ascii_letters

# Add Numbers

if include_numbers == "Y":
    characters += string.digits # .digits = contains numbers from 0 to 9 

# Add Symbols

if include_symbols == "Y":
    characters += string.punctuation # .puntuation = Adding special symbols to the password

# Generating Password

password = ""

for i in range(length):
    password += random.choice(characters)

# Display Password 

print("\nGenerated Password:")
print(password)

# Password Strength Checker (in 3 dimensions)

if length < 8:
    strength = "Weak"

elif length < 12:
    strength = "Medium"

else:
    strength = "Strong"

print("Password Strength:", strength)
```
