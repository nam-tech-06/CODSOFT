```python
# Task 2 = Making Calculator #


# Smart Calculator

history = []
total_calculations = 0

while True:
    
    print("\n=========================================")
    print("           SMART CALCULATOR")
    print("\n=========================================")
    print("1. Addition")
    print("2. Substraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Calculate Square")
    print("6. Calculate Cube")
    print("7. Show History")
    print("8. Clear History")
    print("9. Exit")
    
    choice = input("\nEnter your choice: ")
    
# Addition
    if choice == "1":

        num1 = float(input("Enter First Number: "))
        num2 = float(input("Enter Second Number: "))

        result = num1 + num2
        print("Result = ",result)

        history.append(f"{num1} + {num2} = {result}")
        total_calculations += 1

# Subtraction
    elif choice == "2":

        num1 = float(input("Enter First Number: "))
        num2 = float(input("Enter Second Number: "))

        result = num1 - num2
        print("Result = ",result)

        history.append(f"{num1} - {num2} = {result}")
        total_calculations += 1


# Multiplication
    elif choice == "3":

        num1 = float(input("Enter First Number: "))
        num2 = float(input("Enter Second Number: "))

        result = num1 * num2
        print("Result = ",result)

        history.append(f"{num1} * {num2} = {result}")
        total_calculations += 1


# Division
    elif choice == "4":

        num1 = float(input("Enter First Number: "))
        num2 = float(input("Enter Second Number: "))


        if num2 == 0:
            print("Cannot Divide by zero!")

        else:
            result = num1 / num2
            print("Result = ",result)

        history.append(f"{num1} / {num2} = {result}")
        total_calculations += 1


# Calculate Square
    elif choice == "5":

        num = float(input("Enter First Number: "))

        result = num ** 2
        print("Square = ",result)

        history.append(f"{num}^2 = {result}")
        total_calculations += 1


# Calculate Cube
    elif choice == "6":

        num = float(input("Enter First Number: "))

        result = num ** 3
        print("Cube = ",result)

        history.append(f"{num}^3 = {result}")
        total_calculations += 1


# Show History
    elif choice == "7":
        if len(history) == 0:
            print("\nNo calculations performed yet.")

        else:
            print("\n------ HISTORY ------")

        for item in history:
            print(item)
            print("----------------------")
            print("Total Calculations:", total_calculations)

    
# Clear History
    elif choice == "8":

        history.clear()
        total_calculations = 0

        print("History Cleared Successfully!")


 # Exit
    elif choice == "9":

        print("\nThank You For Using Smart Calculator!")
        break


# Invalid Choice
    else:
        print("Invalid Choice! Please try again.")
```
