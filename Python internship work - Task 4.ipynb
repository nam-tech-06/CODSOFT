```python
# Task 4 = Rock, Paper and Scissors Game #


import random # importing random to generate random choices

print("=================================")
print("    ROCK PAPER SCISSORS GAME")
print("=================================")

choices = ["Rock", "Paper", "Scissors"] # available choices to play the game

user_score = 0
computer_score = 0

while True:
    
    # Display menu options
    print("\n1. Rock")
    print("2. Paper")
    print("3. Scissors")
    print("4. Exit")

    choice = input("\nEnter your choice: ") # user input

    # For Rock
    if choice == "1":
        user_choice = "Rock"
        
    # For Paper
    elif choice == "2":
        user_choice = "Paper"
        
    # For Scissors
    elif choice == "3":
        user_choice = "Scissors"
    
    # Exit and display the final scores of the game
    elif choice == "4":
        print("\nFinal Score")
        print("User Score:", user_score)
        print("Computer Score:", computer_score)
        print("Thank You For Playing!")
        break

    else:
        print("Invalid Choice!") # To handle invalid inputs
        continue

    
        
    computer_choice = random.choice(choices)

    print("\nYou Chose:", user_choice)
    print("Computer Chose:", computer_choice)

    if user_choice == computer_choice: # Checks for Tie
        print("It's a Tie!")
    
    # Checks all the conditions..
    elif (
        (user_choice == "Rock" and computer_choice == "Scissors") or
        (user_choice == "Paper" and computer_choice == "Rock") or
        (user_choice == "Scissors" and computer_choice == "Paper")
    ):
        print("You Win!")
        user_score += 1

    else:
        print("Computer Wins!")
        computer_score += 1

     # Display current scores
    print("\nCurrent Score")
    print("User:", user_score)
    print("Computer:", computer_score)
    
    
```
